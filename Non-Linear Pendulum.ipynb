## Non-linear Pendulum. 

Consider the pendulum equation again, but this time without any approximation. In other words, we will not linearize the differential equation by assuming small amplitude of oscillation. The equation (in radial coordinates) is given
by:
$$ \frac{d^2 \theta}{dt^2} = - \frac{g}{\ell} \sin \theta$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def leapfrog(x0,v0,t0,T,tau,k):   # k=g/l
    dt=tau/10
    v=[v0]
    x=[x0]
    a=[]
    vhalf=[]
    t=np.linspace(0,T,int(T/dt)+1)
    a.append(0)
    a[0]=-k*np.sin(x[0])
    vhalf.append(0)
    vhalf[0]=v[0]+a[0]*dt/2
    for i in range(int(T/dt)):
        a.append(0)
        a[i]=-k*np.sin(x[i])
        vhalf.append(0)
        vhalf[i+1]=vhalf[i]+a[i]*dt
        x.append(0)
        x[i+1]=x[i]+vhalf[i+1]*dt
    x=np.array(x)
    vhalf=np.array(vhalf)    
    a=np.array(a)
    v=vhalf+a*dt/2

    return [x,v,a,t]
```

```python
x0=np.linspace(np.pi/2,np.pi/10,100)
x,v,a,t=leapfrog(np.pi/2,0,0,30,.01,1)
```

```python
for i in x0:
    plt.plot(t,leapfrog(i,0,0,30,.01,1)[0])
plt.xlabel('Amplitude')
plt.ylabel('Time')
```

It is clear from the above figure that the time period of the oscillations with different amplitude is different
